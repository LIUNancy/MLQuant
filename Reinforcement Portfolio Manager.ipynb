```python
from datetime import datetime
import yfinance as yf
start_date = datetime(2010, 1, 1)
end_date = datetime(2020, 12, 31)
df = yf.download('SPY', start=start_date, end=end_date, auto_adjust=False)
```

```python
## Convert the DataFrame index to datetime and add a fixed time offset to every timestamp
```

```python
from datetime import timedelta
import pandas as pd
df.index = pd.to_datetime(df.index) + timedelta(hours=15, minutes=59, seconds=59)
```

```python
# Adjusted Close is the closing price of a stock after accounting for corporate actions
# max_price_scaler: normalizes the price (likely for ML model scaling)
```

```python
max_price_scaler = 5_000.0
max_price_scaler = 1
max_volume_scaler = 1.5e8
df['Open'] = df['Adj Close'] / df['Close'] * df['Open'] / max_price_scaler 
#  adjusts Open prices to match the adjusted price scale
```

```python
df['High'] = df['Adj Close'] / df['Close'] * df['High'] / max_price_scaler
df['Low'] = df['Adj Close'] / df['Close'] * df['Low'] / max_price_scaler
df['Volume'] = df['Adj Close'] / df['Close'] * df['Volume'] / max_volume_scaler
df['Close'] = df['Adj Close'] / max_price_scaler
```

```python
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
df.head()
```

```python
df.columns = [f"{ticker}:{price.lower()}" for price, ticker in df.columns]
df.columns
```

```python
# ses the ta (Technical Analysis) library to compute the MACD indicator from your DataFrame.
import ta
macd = ta.trend.MACD(close=df[f'{'SPY'}:close'])
```

```python
# MACD = Moving Average Convergence Divergence, creates a MACD object
macd
```

```python
df[f'{'SPY'}:macd'] = macd.macd()
df[f'{'SPY'}:macd_diff'] = macd.macd_diff()
df[f'{'SPY'}:macd_signal'] = macd.macd_signal()
```

```python
# computes the Relative Strength Index (RSI), a momentum indicator. Default parameters: window=14 (14-period RSI)
# RSI measures the speed & magnitude of price changes
rsi = ta.momentum.RSIIndicator(close=df[f'{'SPY'}:close'])
```

```python
df[f'{'SPY'}:rsi'] = rsi.rsi()
df['SPY:rsi']
```

```python
# uses the ta (Technical Analysis) library to create a Bollinger Bands indicator object.
# ta.volatility.BollingerBands() computes Bollinger Bands, which are volatility indicators around a moving average.
# Parameters: close= → closing price series
# window=20 → the number of periods for the moving average (default 20)
# window_dev=2 → number of standard deviations for the upper/lower bands (default 2)
bb = ta.volatility.BollingerBands(close=df[f'{'SPY'}:close'], window=20, window_dev=2)
df[f'{'SPY'}:bb_bbm'] = bb.bollinger_mavg()
df[f'{'SPY'}:bb_bbh'] = bb.bollinger_hband()
df[f'{'SPY'}:bb_bbl'] = bb.bollinger_lband()
```

Bollinger Bands consist of three lines:

Middle Band (MB) → Simple Moving Average (SMA) of the close, period = window

Upper Band (UB) → MB + (window_dev × standard deviation)

Lower Band (LB) → MB - (window_dev × standard deviation)

They measure volatility:

Price near upper band → possibly overbought

Price near lower band → possibly oversold

```python
# compute the Average True Range (ATR), a volatility indicator, calculates the True Range (TR) for each period, then smooths it to get ATR.
atr = ta.volatility.AverageTrueRange(high=df[f'{'SPY'}:high'], low=df[f'{'SPY'}:low'], close=df[f'{'SPY'}:close'])
```
